```python
import requests
import time

# URL de l'API Infoclimat permettant de récupérer les données météorologiques.
# Cette API est utilisée pour extraire des observations historiques par station.
url = "https://www.infoclimat.fr/opendata/"

# Liste des stations identifiées comme exploitables après filtrage.
# Ces identifiants correspondent aux capteurs situés à Paris retenus pour le MVP.
liste_stations = ["000BV", "000CT", "000EW", "ME098"]

# Parcours des stations une par une afin de limiter la charge sur l'API.
# Cette approche permet également d'analyser individuellement la structure des données.
for station in liste_stations:
    # Paramètres envoyés à l'API
    # Ils définissent notamment :
        # le format de sortie (ici CSV pour faciliter les contrôles)
        # la période étudiée
        # l'identifiant de la station
        # la clé API nécessaire pour l'authentification
        params = {
            "method": "get",
            "format": "csv", # Le format CSV est choisi pour une lecture rapide et un contrôle manuel.
            "stations[]": station,
            "start": "2026-06-10",
            "end": "2026-06-22",
            "token": "xxxxx" # À sécuriser après notre MVP (ne pas exposer en clair).
        }
        try:
            # Envoi de la requête HTTP vers l'API:
            response = requests.get(url, params=params)

            # Vérification du bon déroulement de la requête :
            if response.status_code == 200:

                # Le contenu de la réponse est récupéré sous forme de texte CSV :
                csv_content = response.text

                # Génération d'un nom de fichier spécifique à la station :
                filename = f"donnees_meteo_infoclimat_{station}.csv"

                # Écriture du contenu dans un fichier CSV local.
                # Ce choix permet de vérifier facilement la qualité et la structure des données.
                with open(filename, mode='w', encoding='utf-8') as file:
                    file.write(csv_content)
                print(f"Succès : Fichier CSV généré et sauvegardé sous '{filename}'.")

                # Affichage des premières lignes pour vérification.
                # Cette étape permet de vérifier rapidement :
                    # La structure des colonnes
                    # La présence des données
                    # La cohérence du format
                print(f"\nAperçu du contenu pour {station}:")
                print("\n".join(csv_content.splitlines()[:5]))
            else:
                # Gestion des erreurs côté API (si code HTTP différent de 200) :
                print(f"Erreur :{station} - {response.status_code} - {response.text}")

        # Gestion des erreurs réseau ou exceptions inattendues :
        except Exception as e:
            print(f"✗ Erreur de connexion pour {station} : {e}")

            # PAUSE DE 5sec Sauf si c'est la dernière station (gagner du temps à la fin).
            # Pause necessaire car la quantité de données récupérable est limité en durée d'exécution (env 3sec).
        if station != liste_stations[-1]:
            print("Pause de 5 secondes.")
            time.sleep(5)
```
