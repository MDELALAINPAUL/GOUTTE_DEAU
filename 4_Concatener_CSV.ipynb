```python
import pandas as pd
import glob
import os

# Cibler des fichiers à traiter.
# Permet de sélectionner  les fichiers CSV générés lors de la phase d'extraction pour l'année étudiée :
annee_extraction = "2025"
# * pour prendre toutes les saisons d'une année :
cible = f"donnees_meteo_*_{annee_extraction}.csv"

# Récupération de la liste des fichiers :
liste_fichiers = glob.glob(cible)

# Vérification de la présence de fichiers :
if not liste_fichiers:
    print("Aucun fichier trouvé correspondant.")
else:
    print(f"{len(liste_fichiers)} fichiers trouvés.")

# Initialisation d'une liste destinée à stocker les dataframes.
# Chaque fichier CSV sera converti en DataFrame avant d'être fusionné.
    dataframes = []

    # Parcours de chaque fichier :
    for fichier in liste_fichiers:
        try:
            # Lecture du CSV avec un délimiteur de type point-virgule :
            df = pd.read_csv(fichier, sep=';', encoding='utf-8')

             # Ajout du dataframe à la liste :
            dataframes.append(df)
            print(f"  - Lu : {fichier} ({len(df)} lignes)")

        except Exception as e:
            # Gestion des erreurs de lecture :
            print(f"  - Erreur lecture {fichier} : {e}")

    # Vérification de la présence de données valides :
    if dataframes:
        # Concaténation de tous les dataframes.
        # ignore_index=True permet de recréer un index unique continu :
        df_final = pd.concat(dataframes, ignore_index=True)

        # Définition du nom du fichier de sortie (variable par année).
        fichier_sortie = f"donnees_meteo_{annee_extraction}_COMPLET.csv"

        # Export du dataframe final vers un fichier CSV.
        # Le séparateur point-virgule est conservé pour cohérence avec les données initiales :
        df_final.to_csv(fichier_sortie, index=False, sep=';', encoding='utf-8')
        
        print(f"\nSuccès : Fichier unique généré -> {fichier_sortie}")
        print(f"Nombre total de lignes : {len(df_final)}")
    else:
        # Message affiché si aucun fichier n'a pu être exploité.
        print("Aucune donnée valide n'a pu être traitée.")
```
