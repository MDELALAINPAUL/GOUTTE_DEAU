```python
import glob

# Définition de l'en-tête standardisé
# Cet en-tête est appliqué à l'ensemble des fichiers afin d'assurer :
    # Une structure homogène des données.
    # Une compatibilité avec les traitements futurs.
    # Une facilitation l'apprentissage.
nouvel_entete = "station_id;dh_utc;temperature_degC;pression_hPa;humidite_%;point_de_rosee_degC;vent_moyen_km/h;vent_rafales_km/h;vent_direction_deg;pluie_3h_mm;pluie_1h_mm\n"

# Cibler des fichiers à traiter.
# Permet de sélectionner automatiquement tous les fichiers CSV générés lors de la phase d'extraction pour l'année étudiée :
annee_extraction = "2025"
# * pour prendre toutes les saisons d'une année :
cible= f"donnees_meteo_*_{annee_extraction}.csv"

# Récupération de la liste des fichiers correspondant à l'année retenue :
liste_fichiers = glob.glob(cible)

print(f"Traitement de {len(liste_fichiers)} fichiers...")
print(f"Actions : Suppression des 6 premières lignes + Insertion du nouvel en-tête.")

# Parcours de chaque fichier pour appliquer les opérations de nettoyage
for fichier in liste_fichiers:
    try:
        # Lecture complète du fichier
        # Cette étape permet d'accéder aux données sous forme de liste de lignes
        with open(fichier, 'r', encoding='utf-8') as f:
            lignes = f.readlines()
        
        # Vérification qu'il y a assez de lignes à supprimer
        if len(lignes) > 6:
            # On garde tout SAUF les 6 premières lignes (index 0 à 5) :
            lignes_conservees = lignes[6:]
            
            # On prépare le nouveau contenu : Nouvel En-tête + Reste des données :
            nouveau_contenu = [nouvel_entete] + lignes_conservees
            
            # Réécriture du fichier avec un écrase si fichier déjà existant :
            with open(fichier, 'w', encoding='utf-8') as f:
                f.writelines(nouveau_contenu)
            
            print(f"{fichier} : Mis à jour avec succès.")
        else:
            print(f"{fichier} : Le fichier contient moins de 6 lignes, ignoré pour éviter la perte de données.")
            
    except Exception as e:
        # Gestion des erreurs éventuelles (lecture/écriture)
        print(f"Erreur sur {fichier} : {e}")

print("Terminé.")
```
