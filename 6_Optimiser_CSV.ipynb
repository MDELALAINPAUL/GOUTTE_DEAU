```python
import pandas as pd

# Définition des fichiers d'entrée et de sortie.
# Le fichier d'entrée correspond au dataset consolidé brut.
# Le fichier de sortie correspondra au dataset nettoyé, prêt pour l'entrainement.
fichier_entree = 'donnees_meteo_2025_COMPLET.csv'
fichier_sortie = 'donnees_meteo_2025_NETTOYEES.csv'

# Chargement du CSV (séparateur point-virgule).
df = pd.read_csv(fichier_entree, sep=';')

# Liste des colonnes à supprimer.
colonnes_a_supprimer = ['pluie_3h_mm', 'vent_rafales_km/h']

# Liste des lignes à supprimer
# Filtrage : On garde uniquement les lignes où station_id N'EST PAS égal à '000BV' et ne sont pas null.
df_filtre = df[
    (df['station_id'] != '000BV') & 
    (df['pluie_1h_mm'].notnull())
]

# Suppression des colonnes :
df_nettoye = df_filtre.drop(columns=colonnes_a_supprimer, errors='ignore')

# Ajout colonne "pleut_il" : "pleut_il" prend la valeur 1 si la pluie sur 1 heure est strictement supérieure à 0, sinon elle prend la valeur 0 :
df_nettoye['pleut_il'] = (df_nettoye['pluie_1h_mm'] > 0).astype(int)

# Suppression de toutes les lignes contenant des valeurs manquantes.
# Le choix a été fait de privilégier la qualité des données plutôt que leur volume :
df_nettoye = df_nettoye.dropna()

# sauvegarde du nouveau fichier:
df_nettoye.to_csv(fichier_sortie, sep=';', index=False)

# Affichage des résultats pour validation :
print(f"Succès : Fichier généré '{fichier_sortie}' sans les colonnes {colonnes_a_supprimer}.")
print(f"Colonnes conservées : {list(df_nettoye.columns)}")
print(f"Nombre de lignes restantes : {len(df_nettoye)}")
```
